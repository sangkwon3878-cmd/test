```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# ðŸ”¹ .env íŒŒì¼ ë¡œë“œ
load_dotenv()

# ðŸ”¹ OpenAI í´ë¼ì´ì–¸íŠ¸ ìƒì„±
client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

# ðŸ”¹ ëŒ€í™” ížˆìŠ¤í† ë¦¬ (system ë©”ì‹œì§€ëŠ” ìµœì´ˆ 1íšŒë§Œ)
messages = [
    {
        "role": "system",
        "content": (
            "ë„ˆëŠ” ì¹œì ˆí•˜ê³  ê°„ê²°í•œ AI ì–´ì‹œìŠ¤í„´íŠ¸ë‹¤. "
            "ìš•ì„¤, í˜ì˜¤ í‘œí˜„, ë¶ˆë²•ì ì¸ ë‚´ìš©ì€ ì ˆëŒ€ ìƒì„±í•˜ì§€ ë§ê³  "
            "í•­ìƒ í•œêµ­ì–´ë¡œë§Œ ëŒ€ë‹µí•´ë¼."
        )
    }
]

print("ðŸ¤– GPTì™€ ëŒ€í™”ë¥¼ ì‹œìž‘í•©ë‹ˆë‹¤. ì¢…ë£Œí•˜ë ¤ë©´ 'exit' ìž…ë ¥")

while True:
    user_input = input("ðŸ™‹ ì‚¬ìš©ìž: ")

    if user_input.lower() == "exit":
        print("ðŸ‘‹ ëŒ€í™”ë¥¼ ì¢…ë£Œí•©ë‹ˆë‹¤.")
        break

    # ðŸ”¹ ì‚¬ìš©ìž ë©”ì‹œì§€ ì¶”ê°€
    messages.append({
        "role": "user",
        "content": user_input
    })

    # ðŸ”¹ GPT í˜¸ì¶œ
    response = client.responses.create(
        model="gpt-4.1-mini",
        input=messages,
        max_output_tokens=300,
        temperature=0.3,
        top_p=0.9
    )

    # ðŸ”¹ GPT ì‘ë‹µ í…ìŠ¤íŠ¸
    assistant_text = response.output_text

    print(f"ðŸ¤– GPT: {assistant_text}")

    # ðŸ”¹ GPT ì‘ë‹µë„ ëŒ€í™” ížˆìŠ¤í† ë¦¬ì— ì €ìž¥
    messages.append({
        "role": "assistant",
        "content": assistant_text
    })

```
